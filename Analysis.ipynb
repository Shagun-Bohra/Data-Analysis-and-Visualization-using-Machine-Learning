### Applied one hot encoding for the column named "CREDDESC" (adopted an inefficient way just to perform clustering.)

```python
data_mapped = data.copy()
data_mapped["CREDDESC"] = data_mapped["CREDDESC"].map({"Bachelors Degree":0, "Associate's Degree":1,
                                                      "Undergraduate Certificate or Diploma":2,
                                                      "Master's Degree":3,
                                                      "Doctoral Degree":4,
                                                      "Graduate/Professional Certificate":5,
                                                      "First Professional Degree":6,
                                                      "Post-baccalaureate Certificate":7})
```

```python
data_mapped.head(10)
```

# Considering CREDDESC and CREDLEV columns for data clustering 

```python
X = data_mapped.iloc[:, [8,7]].values
```

```python
print(X)
```

```python
X.shape
```

# The WCSS Clustering Concept

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

# The Elbow Point Curve

```python
sns.set()
plt.plot(range(1,11), wcss)
plt.title("The Elbow Point Title")
plt.xlabel("No. of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 6, init = 'k-means++', random_state = 0)
Y = kmeans.fit_predict(X)
```

```python
print(Y)
```

# Concatination of Clusters in the given data.

```python
labels = pd.DataFrame(kmeans.labels_)
labeled_data = pd.concat((data_mapped,labels),axis=1)
labeled_data = labeled_data.rename({0:'clusters'},axis=1)
```

```python
labeled_data.head()
```

# Clustering plot with colourful spots (other than black) showing data distribution and black spots showing centroid.

```python
plt.figure(figsize=(5,5))
plt.scatter(X[Y==0,0], X[Y==0,1], s=100, c='green', label='Cluster 1')
plt.scatter(X[Y==1,0], X[Y==1,1], s=100, c='red', label='Cluster 2')
plt.scatter(X[Y==2,0], X[Y==2,1], s=100, c='yellow', label='Cluster 3')
plt.scatter(X[Y==3,0], X[Y==3,1], s=100, c='blue', label='Cluster 4')
plt.scatter(X[Y==4,0], X[Y==4,1], s=100, c='purple', label='Cluster 5')
plt.scatter(X[Y==2,0], X[Y==2,1], s=100, c='Pink', label='Cluster 6')




plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=10, c='black', label='centroids')
plt.title('Cluster')
plt.xlabel('CREDDESC')
plt.ylabel('CREDLEV')
plt.show()
```
